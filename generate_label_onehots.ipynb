```python
#imports
import os
import numpy as np
np.random.seed(1969)
from numpy import genfromtxt
from tqdm import tqdm
from scipy import signal
from glob import glob
import re
import pandas as pd
import gc
from scipy.io import wavfile
import pickle
```

```python
def get_onehot_labels(dirpath, heirarchy, ext='npy',restrictA=False):
    aCount=0
    fpaths = sorted(glob(os.path.join(dirpath, r'*/*' + ext)))
    labels = []
    for fpath in tqdm(fpaths):
        split_by_slash = fpath.split('/')
        speaker = split_by_slash[-2]
        file_name = split_by_slash[-1]
        file_name_split = file_name.split('-')
        syllable = file_name_split[1]
        label = file_name_split[1+heirarchy]
        if(heirarchy==2):
            label = label[:label.find('.')]

        '''if label=='A' and restrictA:
            if aCount>2500:
                continue
            else:
                aCount+=1'''

        labels.append(label)
    print("There are %d samples"%len(labels))

    #convert y to onehot encoded vectors
    n_classes = len(set(labels))
    class_list = list(set(labels)).sort()
    print("There are %s classes. Their distribution:"%n_classes)
    for i in set(labels):
        print("%s : %d"%(i, labels.count(i)))

    onehotdict = {}
    labels = list(set(labels))
    for i in set(labels):
        if i not in onehotdict:
            onehot = [0 for k in range(len(labels))]
            onehot[labels.index(i)] = 1
            onehotdict[i] = onehot

    return onehotdict
```

```python
for i in tqdm(range(1,3)):
    heirarchydict[i] = get_onehot_labels( "/home/rohan/data-telugu/train-mfsc", i)
```

```python
def get_freq(dirpath, heirarchy, ext='npy',restrictA=False):
    aCount=0
    fpaths = sorted(glob(os.path.join(dirpath, r'*/*' + ext)))
    labels = []
    for fpath in tqdm(fpaths):
        split_by_slash = fpath.split('/')
        speaker = split_by_slash[-2]
        file_name = split_by_slash[-1]
        file_name_split = file_name.split('-')
        syllable = file_name_split[1]
        label = file_name_split[1+heirarchy]
        if(heirarchy==2):
            label = label[:label.find('.')]

        '''if label=='A' and restrictA:
            if aCount>2500:
                continue
            else:
                aCount+=1'''

        labels.append(label)
    print("There are %d samples"%len(labels))
    freqdict = {}
    #convert y to onehot encoded vectors
    n_classes = len(set(labels))
    class_list = list(set(labels)).sort()
    print("There are %s classes. Their distribution:"%n_classes)
    for i in set(labels):
        print("%s : %d"%(i, labels.count(i)))
        freqdict[i] = labels.count(i)
    return freqdict
```
